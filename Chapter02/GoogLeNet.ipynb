```python
import torch.nn as nn
```

```python
"""
Parámetro	¿Qué representa?
input_planes	Número de canales de entrada (ej., 192 en im1).
n_channels1x1	Número de filtros en la rama 1x1 conv.
n_channels3x3red	Número de filtros en la reducción (1x1 conv) antes del 3x3 conv.
n_channels3x3	Número de filtros en el 3x3 conv.
n_channels5x5red	Número de filtros en la reducción (1x1 conv) antes del 5x5 conv.
n_channels5x5	Número de filtros en el 5x5 conv (separado en 2 convoluciones 3x3).
pooling_planes	Número de filtros en la rama de max pooling → 1x1 conv.
"""

class InceptionModule(nn.Module):
    def __init__(self, input_planes, n_channels1x1, n_channels3x3red, n_channels3x3, n_channels5x5red, n_channels5x5, pooling_planes):
        super(InceptionModule, self).__init__()
        # 1x1 convolution branch
        self.block1 = nn.Sequential(
            nn.Conv2d(input_planes, n_channels1x1, kernel_size=1),
            nn.BatchNorm2d(n_channels1x1),
            nn.ReLU(True),
        )
 
        # 1x1 convolution -> 3x3 convolution branch
        self.block2 = nn.Sequential(
            nn.Conv2d(input_planes, n_channels3x3red, kernel_size=1),
            nn.BatchNorm2d(n_channels3x3red),
            nn.ReLU(True),
            nn.Conv2d(n_channels3x3red, n_channels3x3, kernel_size=3, padding=1),
            nn.BatchNorm2d(n_channels3x3),
            nn.ReLU(True),
        )
 
        # 1x1 conv -> 5x5 conv branch
        self.block3 = nn.Sequential(
            nn.Conv2d(input_planes, n_channels5x5red, kernel_size=1),
            nn.BatchNorm2d(n_channels5x5red),
            nn.ReLU(True),
            nn.Conv2d(n_channels5x5red, n_channels5x5, kernel_size=3, padding=1),
            nn.BatchNorm2d(n_channels5x5),
            nn.ReLU(True),
            nn.Conv2d(n_channels5x5, n_channels5x5, kernel_size=3, padding=1),
            nn.BatchNorm2d(n_channels5x5),
            nn.ReLU(True),
        )
 
        # 3x3 pool -> 1x1 conv branch
        self.block4 = nn.Sequential(
            nn.MaxPool2d(3, stride=1, padding=1),
            nn.Conv2d(input_planes, pooling_planes, kernel_size=1),
            nn.BatchNorm2d(pooling_planes),
            nn.ReLU(True),
        )
 
    def forward(self, ip):
        op1 = self.block1(ip)  # Rama 1 (1x1 conv)
        op2 = self.block2(ip)  # Rama 2 (1x1 conv → 3x3 conv)
        op3 = self.block3(ip)  # Rama 3 (1x1 conv → 3x3 conv → 3x3 conv)
        op4 = self.block4(ip)  # Rama 4 (3x3 maxpool → 1x1 conv)
        return torch.cat([op1, op2, op3, op4], 1)  # Concatenación a lo largo de los canales
```

```python
import torch
import torch.nn as nn

class GoogLeNet(nn.Module):
    def __init__(self):
        super(GoogLeNet, self).__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(3, 192, kernel_size=3, padding=1),
            nn.BatchNorm2d(192),
            nn.ReLU(True),
        )

        # Bloques de Inception
        self.im1 = InceptionModule(192, 64, 96, 128, 16, 32, 32)
        self.im2 = InceptionModule(256, 128, 128, 192, 32, 96, 64)
        
        self.max_pool = nn.MaxPool2d(3, stride=2, padding=1)
        
        self.im3 = InceptionModule(480, 192, 96, 208, 16, 48, 64)
        self.im4 = InceptionModule(512, 160, 112, 224, 24, 64, 64)
        self.im5 = InceptionModule(512, 128, 128, 256, 24, 64, 64)
        self.im6 = InceptionModule(512, 112, 144, 288, 32, 64, 64)
        self.im7 = InceptionModule(528, 256, 160, 320, 32, 128, 128)

        self.im8 = InceptionModule(832, 256, 160, 320, 32, 128, 128)
        self.im9 = InceptionModule(832, 384, 192, 384, 48, 128, 128)

        self.average_pool = nn.AvgPool2d(7, stride=1)
        self.fc = nn.Linear(1024, 1000)  # 1024 en lugar de 4096

    def forward(self, x):
        x = self.stem(x)
        
        # Primera parte
        x = self.im1(x)
        x = self.im2(x)
        x = self.max_pool(x)
        
        # Segunda parte
        x = self.im3(x)
        x = self.im4(x)
        x = self.im5(x)
        x = self.im6(x)
        x = self.im7(x)
        x = self.max_pool(x)
        
        # Última parte
        x = self.im8(x)
        x = self.im9(x)
        x = self.average_pool(x)
        
        x = torch.flatten(x, 1)  # Equivalente a .view(x.size(0), -1)
        x = self.fc(x)

        return x
```
