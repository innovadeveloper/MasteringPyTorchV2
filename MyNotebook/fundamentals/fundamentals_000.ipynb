## Conceptos generales

---

### Que son las funciones?
Una **funcin** es una relacin matemtica que asocia cada entrada ($x$) con una nica salida ($y$). La notacin general es:
$$
y = f(x)
$$
Aqu:
- $x$ es la **variable independiente** (lo que puedes controlar o medir).
- $y$ es la **variable dependiente** (lo que resulta de aplicar la funcin a $x$).
- $f(x)$ es la regla o frmula que describe cmo se relacionan $x$ e $y$.

**Ejemplo simple**:
Si $f(x) = x^2$, para $x = 2$, la salida ser:
$$
f(2) = 2^2 = 4
$$

---

### Que son las derivadas?
La derivada de una funcin mide cmo cambia su salida $y$ con respecto a su entrada $$. 

Matemticamente, es la **tasa de cambio instantnea de la funcin en un punto especfico**.

#### Ejemplo Prctico de las derivadas : Cambio de velocidad con aceleracin constante


Vamos a usar una funcin que relaciona la **distancia recorrida** con el **tiempo** para explicar qu es la derivada y cmo se aplica.

**1. La frmula que usaremos**
La distancia recorrida ($d$) por un auto que acelera uniformemente se puede modelar como:
$$
d(t) = v_0 \cdot t + \frac{1}{2} \cdot a \cdot t^2
$$
Donde:
- $d(t)$: Distancia recorrida en funcin del tiempo ($t$).
- $v_0$: Velocidad inicial del auto (en metros por segundo, m/s).
- $a$: **Aceleracin constante** del auto (en m/s).
- $t$: Tiempo transcurrido (en segundos, s).

**2. Qu significa esta frmula?**
- El primer trmino ($v_0 \cdot t$) es la distancia recorrida por el auto si se moviera a una velocidad constante $v_0$.
- El segundo trmino ($\frac{1}{2} \cdot a \cdot t^2$) representa el efecto de la aceleracin en la distancia recorrida.

---

**3. Ejemplo con valores**
Supongamos:
- Velocidad inicial: $v_0 = 5 \, \text{m/s}$.
- Aceleracin: $a = 2 \, \text{m/s}^2$.
- Tiempo: $t = 3 \, \text{s}$.

Usamos la frmula para calcular la distancia:
$$
d(3) = 5 \cdot 3 + \frac{1}{2} \cdot 2 \cdot 3^2
$$
$$
d(3) = 15 + \frac{1}{2} \cdot 2 \cdot 9
$$
$$
d(3) = 15 + 9 = 24 \, \text{m}
$$

Esto significa que, despus de 3 segundos, el auto ha recorrido 24 metros.

---

**4. Qu es la derivada en este contexto?**

La derivada de $d(t)$ con respecto a $t$ nos dice **la velocidad instantnea** del auto en cualquier momento:
$$
v(t) = d'(t) = \frac{\text{d}d(t)}{\text{d}t}
$$

Calculemos la derivada de $d(t)$:
$$
d(t) = v_0 \cdot t + \frac{1}{2} \cdot a \cdot t^2
$$
La derivada es:
$$
d'(t) = v_0 + a \cdot t
$$

Aqu:
- $v(t)$: Velocidad del auto en el instante $t$.
- $v_0$: Velocidad inicial.
- $a \cdot t$: Incremento en la velocidad debido a la aceleracin.

**Clculo de la velocidad**
Con los valores anteriores ($v_0 = 5$, $a = 2$):
$$
v(t) = 5 + 2 \cdot t
$$

En $t = 3 \, \text{s}$, la velocidad es:
$$
v(3) = 5 + 2 \cdot 3 = 5 + 6 = 11 \, \text{m/s}
$$

Esto significa que, despus de 3 segundos, el auto se est moviendo a una velocidad de 11 metros por segundo.


**Nota**
Si la **aceleracin fuera variable**, la velocidad se obtiene integrando la aceleracin respecto al tiempo.
La **velocidad instantnea** en cualquier tiempo ** es el resultado de esa integral.
La **aceleracin instantnea** se obtiene tomando la derivada de la velocidad con respecto al tiempo.

---

###Sabas que ...?

La integral "deshace" la derivada, y la derivada "deshace" la integral. Pero incluso si una de las operaciones es de algn modo "deshecha" por la otra, siempre hay una "constante de integracin" que queda pendiente en el caso de la integracin, y una constante arbitraria cuando se deriva. Es como si te devolvieras al punto de inicio pero con una pequea flexibilidad extra.
